```python
bins = [50,60,70,80,90,100]
plt.hist(fifa_new['Overall'], bins=bins, color= '#8344f2')
plt.ylabel('Number of players')
plt.xlabel('Overall Skill Level')
```

```python
import seaborn as sns
```

```python
corMat= fifa_new.corr()
sns.heatmap(corMat, annot=True)   #OMG oopsie
```

```python
plt.figure(figsize=(16, 10))
sns.scatterplot(data=fifa_new, x='Penalties', y='Potential' ,alpha=0.7)   #eww :)
```

```python
def stripvalues(feature,word):
  fifa_new[feature]= [int(x.strip(word)) if type(x)==str else x for x in fifa_new[feature]]
```

```python
stripvalues('Weight', 'lbs')
```

```python
fifa_new['Weight']
```

```python
light = fifa_new.loc[fifa_new['Weight'] < 125 ].count()[0]
light_med = fifa_new.loc[(fifa_new['Weight'] >= 125)  & (fifa_new['Weight'] < 150)].count()[0]
med = fifa_new.loc[(fifa_new['Weight'] >= 150 ) & (fifa_new['Weight'] < 175)].count()[0]
med_heavy  = fifa_new.loc[(fifa_new['Weight'] >= 175) & (fifa_new['Weight'] < 200)].count()[0]
heavy = fifa_new.loc[fifa_new['Weight'] >= 200].count()[0]
```

```python
weights=[light,light_med,med,med_heavy,heavy]
labels = ['under 125', '125-150','150-175', '175-200', 'above 200']
explode = (.4, .1,0,0,.4)
plt.figure(figsize=(10,15))
plt.style.use('ggplot')
plt.title( ' Weight(lbs) Distribution of FIFA players ')
plt.pie(weights, labels=labels, autopct= '%0.2f', explode= explode)
```

```python
barca = fifa_new.loc[fifa_new['Club'] == 'FC Barcelona']['Overall']
madrid = fifa_new.loc[fifa_new['Club'] == 'Real Madrid']['Overall']
revs = fifa_new.loc[fifa_new['Club'] == 'New England Revolution']['Overall']
```

```python
labels= ['FC Barcelona', 'Real Madrid', 'New England Revolution']
plt.figure(figsize=(7,7))
plt.title('FIFA teams comparision')
boxes = plt.boxplot([barca,madrid, revs], labels=labels)
for box in boxes['boxes']:
  box.set(color='blue')
plt.show()
```
